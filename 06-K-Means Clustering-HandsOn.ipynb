# K-Means Clustering

## Method Used

K Means Clustering is an unsupervised learning algorithm that tries to cluster data based on their similarity. Unsupervised learning means that there is no outcome to be predicted, and the algorithm just tries to find patterns in the data. In k means clustering, we have the specify the number of clusters we want the data to be grouped into. The algorithm randomly assigns each observation to a cluster, and finds the centroid of each cluster. Then, the algorithm iterates through two steps:
Reassign data points to the cluster whose centroid is closest. Calculate new centroid of each cluster. These two steps are repeated till the within cluster variation cannot be reduced any further. The within cluster variation is calculated as the sum of the euclidean distance between the data points and their respective cluster centroids.

### Import Libraries and Get Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# In Unsupervised learning we aren't trying to predict, but find patterns in the data
from sklearn.datasets import make_blobs #used to generate artifical data in form of clusters
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape #[0] = First element of this tuple
```

### Vizualize some data

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow') #Comparing blobs all rows of 1st col. and all rows of 2nd col.
```

```python
data[1] # Has 4 centers to this cluster
#Given Labels for element1
```

### Use K-Means Algorithm -Clusters Creation

```python
from sklearn.cluster import KMeans
```

* It randomly starts to assign each observation to a cluster and then finds centroid of each cluster
* Then it will reassign the data points to the cluster whose centroid is closest
* Calculates new center for each cluster and then repeats it over again until cluster variation can't be reduced any further.


```python
kmeans = KMeans(n_clusters=4) #Here as it is a data set built by us, we can predict cluster no.
```

```python
kmeans.fit(data[0]) # Data[0] = feature
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_ #Predicted Labels
```

```python
# In real world data, it is not possible to predict labels appropiately
```

```python
#Comparing Original and KMeans clustered data using Tuple Unpacking

fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

#In Plot-1 the clusters are labelled acc. KMeans Algorithm for Original Data
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')


#In Plot-2, the clusters are labelled on initial stage for Original Data
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
